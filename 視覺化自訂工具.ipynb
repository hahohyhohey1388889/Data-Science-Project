# 分配圖

## 單一分佈圖

```python
def single_distribution(df, feature):    
    sns.set_style('whitegrid')
    plt.figure(figsize=(16,6))
    sns.kdeplot(df[feature], bw=0.5, color="r")
    plt.xlabel(feature, fontsize=15)
    plt.tick_params(axis='x', which='major', labelsize=15, pad=8)
    plt.tick_params(axis='y', which='major', labelsize=15)
    plt.show()
```

```python
single_distribution(dataframe, 'Alcohol')
```

## 多個分佈圖

```python
def Mult_distribution(df, cols):    
    sns.set_style('whitegrid')
    plt.figure()
    fig, ax = plt.subplots(5,3,figsize=(18,22))
    interval = 0
    for col in cols:
        interval += 1
        plt.subplot(5,3,interval)
        sns.kdeplot(df[col], bw=0.5, color="r")
        plt.xlabel(col, fontsize=8)
        locs, labels = plt.xticks()####?
        plt.tick_params(axis='x', which='major', labelsize=6, pad=8)
        plt.tick_params(axis='y', which='major', labelsize=6)
    plt.show()
```
